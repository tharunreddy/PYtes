```python
import requests
import functools
import json
```

```python
URL = "https://cricketapi.platform.iplt20.com//fixtures/{match_id}"
```

```python
@functools.lru_cache()
def get_match(match_id):
    r = requests.get(URL.format(match_id=match_id))
    return r.json()
```

```python
IPL_2018_MATCHES = range(7894, 7950)
```

```python
TEAM_MAPPINGS = {
    1: 'chennai-super-kings',
    3: 'delhi-daredevils',
    4: 'kings-xi-punjab',
    5: 'kolkata-knight-riders',
    6: 'mumbai-indians',
    62: 'sunrisers-hyderabad',
    8: 'rajasthan-royals',
    9: 'royal-challengers-bangalore'
}
```

```python
matches = []
for match_id in IPL_2018_MATCHES:
    match = get_match(match_id)
    
    matches.append(
        {
            'starttime': match['timestamp'],
            'team1': TEAM_MAPPINGS.get(int(match['scheduleEntry']['team1']['team']['id'])),
            'team2': TEAM_MAPPINGS.get(int(match['scheduleEntry']['team2']['team']['id']))
        }
    )
    
    
    
    
```
