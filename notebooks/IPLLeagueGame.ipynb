```python
import requests
import arrow
```

```python
URL = 'https://2fjfpxrbb3.execute-api.ap-southeast-1.amazonaws.com/production/leagueapi/getleaguemembers?leagueId=ip3NjxML'
userid = 'tarunreddy.bethi@gmail.com'

def get_league_data():
    r = requests.get(URL, headers={'userid': userid})
    return r.json()['data']

league_data = get_league_data()
```

```python
members = league_data['leagueMembers']
#Adding a noop to make even number of teams
teams = [member['teamName'] for member in members] + ['NoOP'] 
teams
```

```python
def schedule(teams):
    """Generates a round robin schedule"""
    teams = teams.copy()
    while True:
        mid = len(teams)//2
        for match in zip(teams[0:mid], teams[mid:]):
            yield match
        last_team = teams.pop()
        teams.insert(1, last_team)
```

```python
now = arrow.Arrow.now()

# Get next 500 matches according to roundrobin schedule
g = schedule(teams)
matches = list(((next(g) for _ in range(500))))

def format(matches):
    for match in matches:
        print("{} vs {}".format(match[0], match[1]))
        
# Print matches for next 42 days of IPL        
for i in range(1, 42):
    time = now.replace(days=i)
    print(time.date())
    format(matches[i*5: i*5 + 5])
    print()
    
    
    
    
```
