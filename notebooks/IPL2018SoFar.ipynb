```python
import json
from collections import Counter
from tabulate import tabulate

import pandas as pd
```

```python
HISTORY_DATA = json.load(open('data/teams_squad_history.json'))
```

## Who is the most picked player in our league? 

```python
players = []
for team, history in HISTORY_DATA.items():
    for squad in history:
        players.extend(squad['players'])
    
counter = Counter(players)

print(tabulate(counter.most_common()))
```
